```python
# Examples from
#https://seaborn.pydata.org/generated/seaborn.tsplot.html

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(22)
sns.set(color_codes=True)

x = np.linspace(0, 15, 31)
print('x = {}\n'.format(x))

data = np.sin(x) + np.random.rand(10, 31) + np.random.randn(10, 1)
print('data = {}'.format(data))
```

```python
ax = sns.tsplot(data=data) # translucent confidence bars
```

```python
gammas = sns.load_dataset("gammas")
print(gammas)
```

```python
# dataframe input with multiple categories (conditions)
ax = sns.tsplot(time="timepoint", value="BOLD signal", unit="subject", condition="ROI", data=gammas)
```

```python
ax = sns.tsplot(data=data, err_style="ci_bars", color="g") # error bars at points of observation
```

```python
ax = sns.tsplot(data=data, err_style="ci_bars", interpolate=False) # don't interpolate
```

```python
ax = sns.tsplot(data=data, ci=[68, 95], color="m") # several confidence bands
```

```python
ax = sns.tsplot(data=data, ci="sd") # standard deviation
```

```python
ax = sns.tsplot(data=data, estimator=np.median)
```

```python
ax = sns.tsplot(data=data, err_style="boot_traces", n_boot=500)
```

```python
ax = sns.tsplot(data=data, err_style="unit_traces")
```
